```python
import os
#import tweepy
import re
import string
from textblob import TextBlob
import preprocessor as p
#from preprocessor.api import clean, tokenize, parse
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import datetime
from datetime import timedelta
import numpy as np
import emoji
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
```

```python
def remove_pattern(text, pattern_regex):
    r = re.findall(pattern_regex, text)
    for i in r:
        text = re.sub(i, '', text)
    return text

data_tweet_cleaning_gedebage3['clean_tweet'] = np.vectorize(remove_pattern)(data_tweet_cleaning_gedebage3['text'], " *RT* | *@[\w]*")
data_tweet_cleaning_gedebage3.head()
```

```python
# menghilangkan simbol dan http
def remov(text):
    text = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",text).split())
    return text
data_tweet_cleaning_gedebage3['remove_http'] = data_tweet_cleaning_gedebage3['clean_tweet'].apply(lambda x: remov(x))
data_tweet_cleaning_gedebage3.head(10)
```

```python
#Menghilangkan hastag
def remov(tweet):
    #hilangkan stock market ticker like $GE
    tweet = re.sub(r'\$\w*', '', tweet)
    
    #hlangkan old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    
    #hilangkan hashtags
    tweet = re.sub(r'#', '', tweet)
    
    #hilangkan angka
    tweet = re.sub('[0-9]+', '', tweet)
        
    return tweet
data_tweet_cleaning_gedebage3['remove_hastag'] = data_tweet_cleaning_gedebage3['remove_http'].apply(lambda x: remov(x))
data_tweet_cleaning_gedebage3.head()
```

```python
#Menghilangkan Data duplikat
data_tweet_cleaning_gedebage3.drop_duplicates(subset = 'remove_hastag', keep = 'first', inplace = True)
data_tweet_cleaning_gedebage3.head(10)
```

```python
# cek ulang data duplikat
data_tweet_cleaning_gedebage3.describe()
```

```python
#Membuat tweet menjadi kata dasar & Menghapus kata tidak baku
#stopword

nltk.download('stopwords')

stopwords_indonesia = stopwords.words('indonesian')

stop_factory = StopWordRemoverFactory().get_stop_words()
more_stopwords = [
    'yg', 'utk', 'cuman', 'deh', 'Btw', 'tp', 'tapi', 'gua', 'gue', 'gw', 'lo', 'lu', 'kalo', 'klo', 'trs', 'jadi',
    'jd', 'nih', 'ntar', 'nya', 'lg', 'gak', 'gk', 'ngga', 'gak', 'dpt', 'dr', 'kpn', 'emng', 'kyk', 'kok', 'donk',
    'dong', 'yah', 'u', 'yah', 'gak', 'km', 'eh', 'lah', 'nah', 'mas', 'sob', 'bro', 'thx', 'ato', 'or', 'rp', 'jt',
    'rb', 'gpp', 'sm', 'usah', 'moga', 'gaes', 'guys', 'coy', 'kmrn', 'nemu', 'yukk', 'wkwkwk', 'hehehe', 'hahaha',
    'hehe', 'haha', 'lbh', 'cpt', 'cpet', 'ku', 'mba', 'mas', 'mz', 'sdh', 'sdg', 'ih', 'bs', 'krn', 'jgn', 'sapa',
    'spt', 'wakaka', 'mah', 'cm', 'byr', 'byar', 'kzl', 'dlu', 'ciyee', 'ga'
]
data = stop_factory + more_stopwords

dictionary = ArrayDictionary(data)
str = StopWordRemover(dictionary)

print(data)

factory = StemmerFactory()
stemmer = factory.create_stemmer()

#tokenize
from nltk.tokenize import TweetTokenizer

#emoticons happy
emoticons_happy =set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8]', '=)',
    ':}', ':>)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD',
    '=-D', '=D', '=-3', 'x-p', 'XP', ':-p', ':p', '=p', ':-b'
])

#emoticons_sad
emoticons_sad = set([
    ':-(', ':(', ';(', ':o(', ':[', ':3', ':c(', ':>', '=[', '8[', '=(',
    ':{', '>:(', 
])

#all emoticons
emoticons = emoticons_happy.union(emoticons_sad)

def clean_tweets(tweet):
    
    #tokenize tweets
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, reduce_len=True)
    tweet_tokens = tokenizer.tokenize(tweet)
    
    tweets_clean = []
    for word in tweet_tokens:
        if(word not in data and
          word not in emoticons and
          word not in string.punctuation):
        #tweets_clean.append(word)
            stem_word = stemmer.stem(word)
            tweets_clean.append(stem_word)
        
    return tweets_clean
data_tweet_cleaning_gedebage3['Tweety'] = data_tweet_cleaning_gedebage3['remove_hastag'].apply(lambda x: clean_tweets(x))
```

```python
#menghilangkan punctuation
def remove_punct(text):
    text = " ".join([char for char in text if char not in string.punctuation])
    return text
data_tweet_cleaning_gedebage3['tweet_clean'] = data_tweet_cleaning_gedebage3['Tweety'].apply(lambda x: remove_punct(x))
data_tweet_cleaning_gedebage3.head(10)
```
